#### Kryptoarithmetik

Regeln:
*   Jeder Buchstabe steht für eine andere Ziffer.
*   Der erste Buchstabe jedes Wortes darf nicht 0 sein.
*   Die arithmetische Gleichung muss korrekt sein.

```python
''' A1
Löse folgendes kryptoarithmetische Rätsel 

      S E N D
    + M O R E
    -----------
    = M O N E Y

'''
```

```python
include "alldifferent.mzn";

var 1..9: S;
var 0..9: E;
var 0..9: N;
var 0..9: D;
var 1..9: M;
var 0..9: O;
var 0..9: R;
var 0..9: Y;

constraint           1000 * S + 100 * E + 10 * N + D
                   + 1000 * M + 100 * O + 10 * R + E
       = 10000 * M + 1000 * O + 100 * N + 10 * E + Y;

constraint alldifferent([S,E,N,D,M,O,R,Y]);

solve satisfy;

output ["   \(S)\(E)\(N)\(D)\n",
        "+  \(M)\(O)\(R)\(E)\n",
        "= \(M)\(O)\(N)\(E)\(Y)\n"];
```

```python
''' A2
Wieviele Lösunge gibt es für das folgende kryptoarithmetische Rätsel 

      C U H K
    + M E L B
    -----------
    = L O R E 

'''
```

```python
var 1..9: C;
var 0..9: U;
var 0..9: H;
var 0..9: K;
var 1..9: M;
var 0..9: E;
var 1..9: L;
var 0..9: B;
var 0..9: O;
var 0..9: R;

constraint
           1000 * C + 100 * U + 10 * H + K
         + 1000 * M + 100 * E + 10 * L + B
         = 1000 * L + 100 * O + 10 * R + E;

include "alldifferent.mzn";
constraint alldifferent([C,U,H,K,M,E,L,B,O,R]);

solve satisfy;

output
["  \(C) \(U) \(H) \(K)\n" ++
 "+ \(M) \(E) \(L) \(B)\n" ++
 "---------\n" ++
 "= \(L) \(O) \(R) \(E)\n"];
```

```python
''' A3
Ermittle eine Lösung für folgendes kryptoarithmetische Rätsel 

      B E A N S
    + B E A N S
    -----------
  = C R Y I N G

'''
```

```python
var 1..9: B; var 0..9: E; var 0..9: A; var 0..9: N; var 0..9: S;
var 1..9: C; var 0..9: R; var 0..9: Y; var 0..9: I; var 0..9: G;

constraint    10000 * B + 1000 * E + 100 * A + 10 * N + S
            + 10000 * B + 1000 * E + 100 * A + 10 * N + S
            = 100000 * C + 10000 * R + 1000 * Y + 100 * I + 10 * N + G;

include "alldifferent.mzn";
constraint alldifferent ([B,E,A,N,S,C,R,Y,I,G]);

solve satisfy;
output

["    \(B) \(E) \(A) \(N) \(S)\n" ++
 "+   \(B) \(E) \(A) \(N) \(S)\n" ++
 "-------------\n" ++
 "= \(C) \(R) \(Y) \(I) \(N) \(G)\n"];
```

```python
''' A4
Ermittle eine Lösung für folgendes kryptoarithmetische Rätsel 

      C A O
    + Z H I
    + C A O 
    +   P I
    -----------
  = T E S T

Zusätzlich muss gelten: P ist dreimal Z und O ist 0.
Wieviele Lösungen hat dieses Problem?
'''
```

```python
var 1..9: C; var 0..9: A; var 0..9: O;
var 1..9: Z; var 0..9: H; var 0..9: I;
var 1..9: P; var 1..9: T; var 0..9: E; var 0..9: S;

constraint    100 * C + 10 * A + O
            + 100 * Z + 10 * H + I
            + 100 * C + 10 * A + O
            + 10 * P + I
            = 1000 * T + 100 * E + 10 * S + T;
            
constraint P = 3*Z;
constraint O = 0;

include "alldifferent.mzn";
constraint alldifferent ([C,A,O,Z,H,I,P,T,E,S]);

solve satisfy;

output
["  \(C) \(A) \(O)\n" ++
 "+ \(Z) \(H) \(I)\n" ++
 "+ \(C) \(A) \(O)\n" ++
 "  + \(P) \(I)\n" ++
 "---------\n" ++
 "= \(T) \(E) \(S) \(T)\n"];
```

```python
''' A5
Ermittle eine Lösung für folgendes kryptoarithmetische Rätsel.
Der Wert der Zahl, die durch VERSE dargestellt wird, soll möglichst groß sein.

      S E V E N
    + P A C E S
    -----------
    = V E R S E

'''
```

```python
var 1..9: S; var 0..9: E; var 1..9: V; var 0..9: N;
var 1..9: P; var 0..9: A; var 0..9: C; var 0..9: R;
var int: result = 10000 * V + 1000 * E + 100 * R + 10 * S + E;

constraint
              10000 * S + 1000 * E + 100 * V + 10 * E + N
            + 10000 * P + 1000 * A + 100 * C + 10* E + S
            = result;

include "alldifferent.mzn";

constraint alldifferent ([S,E,V,N,P,A,C,R]);

solve maximize result;

output
["  \(S) \(E) \(V) \(E) \(N)\n" ++
 "+ \(P) \(A) \(C) \(E) \(S)\n" ++
 "-----------\n" ++
 "= \(V) \(E) \(R) \(S) \(E)\n"];

```

```python
''' A6
Die Starttemperatur ist 25. Die Gebäude Soll-Temperatur wird nacheinander eingestellt auf
folgende Werte [35, 35, 20, 20, 20]. Die neue Gebäude Ist-Temperatur ist der abgerundete Mittelwert
zwischen der neuen Soll-Temperatur und der alten Ist-Temperatur. Berechne die Abfolge der Ist-Temperaturen.
'''

int: start = 25;
array[1..5] of int: soll = [35, 35, 20, 20, 20];

array[0..5] of var int: ist;           % decision Variables

constraint ist[0] = start;

constraint
  forall(i in 1..5)(
    ist[i] = floor((ist[i-1] + soll[i]) / 2)
  );

solve satisfy;

output ["Ist-Temperaturen: \(ist)"];
```
