```python
import matplotlib.pyplot as plt
import numpy as np    
```

```python
def show(ax, f, von, bis, label):
    
    ax.axvline(0,color='black',linewidth=1)
    ax.axhline(0,color='black',linewidth=1)
    X = np.linspace(von,bis,500)
    ax.grid(True)
    Y = f(X)
    ax.plot(X,Y,label=label)  
    ax.legend()
 
def f1(x):
    return np.exp(x)

def f2(x):
    return np.exp(-x)

def f3(x):
    return -np.exp(x)

def f4(x):
    return -np.exp(-x)


fig, axs = plt.subplots(2,2,figsize=(8,5))
 
show(axs[0][0],f1,-3,2,'$e^{x}$')
show(axs[0][1],f2,-2,3,r'$e^{-x}$')
show(axs[1][0],f3,-3,2,r'$-e^{x}$')
show(axs[1][1],f4,-2,3,r'$-e^{-x}$')
plt.show()
```

```python
def f1(x):
    return x*np.exp(x)

def f2(x):
    return x*np.exp(-x)

def f3(x):
    return x*-np.exp(x)

def f4(x):
    return x*-np.exp(-x)


fig, axs = plt.subplots(2,2,figsize=(8,5))
 
show(axs[0][0],f1,-3,1,r'$x \cdot e^{x}$')
show(axs[0][1],f2,-1,3,r'$x \cdot e^{-x}$')
show(axs[1][0],f3,-3,1,r'$-x \cdot e^{x}$')
show(axs[1][1],f4,-1,3,r'$-x \cdot e^{-x}$')
plt.show()
```

```python
def f1(x):
    return x**2*np.exp(x)

def f2(x):
    return x**2*np.exp(-x)

def f3(x):
    return x**2*-np.exp(x)

def f4(x):
    return x**2*-np.exp(-x)


fig, axs = plt.subplots(2,2,figsize=(8,5))
 
show(axs[0][0],f1,-3,1,r'$x^2 \cdot e^{x}$')
show(axs[0][1],f2,-1,3,r'$x^2 \cdot e^{-x}$')
show(axs[1][0],f3,-3,1,r'$-x^2 \cdot e^{x}$')
show(axs[1][1],f4,-1,3,r'$-x^2 \cdot e^{-x}$')
plt.show()
```

#### Deich

Eine Deich knnen wir mit der Funktion $f(x) = a x^2 \cdot e^{bx}$ modellieren.


```python
def deich(ax, a, b, von, bis):

    def f(x,a,b):
        return a*x**2*np.exp(b*x)

    a0 =  '' if a == 1 else str(a)
    b0 =  '' if b == 1 else str(b)
        
    label=a0+r'$x^2 \cdot e^{'+b0+'x}$'
    ax.axvline(0,color='black',linewidth=1)
    ax.axhline(0,color='black',linewidth=1)
    X = np.linspace(von,bis,500)
    ax.grid(True)
    Y = f(X,a,b)
    ax.plot(X,Y,label=label)  
    ax.legend()
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,3))

deich(axs[0],5,0.5,-20,0)
deich(axs[0],5,0.4,-20,0)
deich(axs[0],5,0.3,-20,0)
deich(axs[1],5,0.5,-20,0)
deich(axs[1],4,0.5,-20,0)
deich(axs[1],3,0.5,-20,0)

plt.show()
```

#### Berg

Einen Berg knnen wir mit der Funktion $f(x) = e^{a(b-x^2)}$ modellieren.

```python
def berg(ax, a, b, von, bis):

    def f(x,a,b):
        return np.exp(a*(b-x**2))

    a0 =  '' if a == 1 else str(a)
    b0 =  '' if b == 1 else str(b)
        
    label=r'$e^{'+a0+'('+b0+'-x^2)}$'
    ax.axvline(0,color='black',linewidth=1)
    ax.axhline(0,color='black',linewidth=1)
    X = np.linspace(von,bis,500)
    ax.grid(True)
    Y = f(X,a,b)
    ax.plot(X,Y,label=label)  
    ax.legend()
```

```python
fig, axs = plt.subplots(1,3,figsize=(14,3))

berg(axs[0],0.5,0,-5,5)
berg(axs[0],0.5,1,-5,5)
berg(axs[0],0.5,2,-5,5)
 
berg(axs[1],0.5,1,-5,5)
berg(axs[1],0.2,1,-5,5)
berg(axs[1],0.1,1,-5,5)

berg(axs[2],0.5,2,-5,5)
berg(axs[2],0.2,5,-5,5)
berg(axs[2],0.1,10,-5,5)

plt.show()
```

#### Fieber

Einen Fieberverlauf knnen wir mit der Funktion $f(x) = ax \cdot e^{-bx}$ modellieren.

```python
def fieber(ax, a, b, von, bis):

    def f(x,a,b):
        return a*x*np.exp(-b*x)

    a0 =  '' if a == 1 else str(a)
    b0 =  '' if b == 1 else str(b)
    label=a0+'x$e^{-'+b0+'x}$'
    ax.axvline(0,color='black',linewidth=1)
    ax.axhline(0,color='black',linewidth=1)
    X = np.linspace(von,bis,500)
    ax.grid(True)
    Y = f(X,a,b)
    ax.plot(X,Y,label=label)  
    ax.legend()
```

```python
fig, axs = plt.subplots(1,2,figsize=(14,3))

fieber(axs[0],1,0.1,0,50)
fieber(axs[0],0.5,0.1,0,50)
fieber(axs[0],0.2,0.1,0,50)

fieber(axs[1],1,0.1,0,50)
fieber(axs[1],1,0.2,0,50)
fieber(axs[1],1,0.3,0,50)
```
