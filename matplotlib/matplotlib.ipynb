Meistens nutzen wir Funktionen aus dem pyplot submodule, das üblicherweise mit dem alias plt importiert wird.

````python
import matplotlib.pyplot as plt
````

#### Einige Grundbegriffe



<img src='figure.png' width='650'>

Der Bereich, in dem gezeichnet wird, heißt *figure*. Innerhalb des *figure* gibt es ein oder mehrere *axes*. Das *axis*-Objekt ist ein zweidimensionales Objekt, das die x und y-Achse enthält. Jede Achse eines *axis* Objekts kann *major* und *minor*-ticks enthalten. Diese ticks können gelabelt oder nicht gelabelt sein. Auch die Achsen selbst können gelabelt werden. Dann können ein oder mehrere plots dem axis-Objekt zugeordnet sein.
Die *legend* zeigt die *lables* der *plots*. Jedes axis-Objekt kann mit einem *title* versehen werden.

````python
plt.subplots()
````

*plt.subplots* returned ein Tuple aus figure und einer axis und gibt die *figure* aus. Da zu dem axis-Objekt noch kein plot hinzugefügt wurde,
ist es im wesentlichen ein leerer Bereich.

````python
fig, ax = plt.subplots()
````

````python
X = [1, 2, 4]
Y = [1, 10, 6]
ax.plot(X, Y, label='Linie')  
fig
````

Weil die *figure* beim Erzeugen immer anzeigt wird, ergänzt man üblicherweise auch die anderen Anweisungen innerhalb einer Zelle. Damit die letzte Zeile der Zelle keine Ausgabe erzeugt, schließt man sie mit ; ab.

````python
fig, ax = plt.subplots()
ax.plot([1,2,3],[1, 5, 2], label='Linie')  
ax.set_title('Ax-Title')
fig.legend();
````

````python
fig, ax = plt.subplots(figsize=(7,3))    
ax.set_title("Title") 
ax.plot([1,2,3],[1, 5, 2], label='Linie')  
ax.plot([1,2.5,3],[3, 0, 4], label='Linie2')  
ax.set_xlabel('x-Achse')
ax.set_ylabel('y-Achse')
fig.legend();
````

#### Multiple Axes

Matplotlib benutzt ein Grid-Layout, um mehrere axes in einem figure zu zeigen.

````python
fig, axs = plt.subplots(3,2)       # rows, columns
axs
````

````python
fig, axs = plt.subplots(3,2,figsize=(10, 6))      # (rows, columns)
ax = axs[1][0]
ax.plot([1,2,3],[1, 5, 2], label='Linie')  
ax.plot([1,2.5,3],[3, 0, 4], label='Linie2')  
ax.set_title('Title')
ax.set_xlabel('x-Achse')
ax.set_ylabel('y-Achse')
fig.legend();
````

#### Punkte plotten

Mit der Funktion *scatter()* können wir Punkte plotten.
Die x-Koordinaten und y-Koordinaten werden als Listen eingegeben. Die Parameter für Dicke oder Farbe können entweder für alle Punkte
oder mit einer Liste für jeden Punkt spezifiziert werden.    

````python
plt.subplots(figsize=(5,3))     
X = [1, 2, 3, 4]
Y = [1, 4, 9, 16]
C = ['red' if y < 5 else 'blue' for y in Y]
S = [30 if y < 12 else 10 for y in Y]
plt.scatter(X, Y, color=C, s=S, marker='o')        # s = dicke
plt.show()
````

Als Farben können wir die üblichen englischen Namen für Farben verwenden. <br>
Als Marker können wir verwenden, z.B:

    'o'  für dicke Punkte
    '.'  für dünne Punkte
    'x'  für eine x-Markierung

````python
X = [1, 2, 3, 4]
Y = [1, 4, 9, 16]
C = ['red','red','blue','blue']
 

plt.figure(figsize=(4,3))                           # 4 inch breit, 3 inch hoch 
plt.scatter(X, Y, color=C, s=30, marker='o')        # s = dicke

plt.show()
````

Die Markierungen der Punkte und die Farben können wir ändern

#### Linien zwischen Punkten

Die Funktion *plot* zeichnet Linien zwischen den Punkten.   

````python
plt.subplots(figsize=(5,3))     
X = [1, 2, 3, 4]
Y = [1, 4, 9, 16]
plt.plot(X, Y, color='green', linestyle=':', marker='x')        # s = dicke
plt.show()
````

Der Parameter *linestyle* bestimmt die Art der Linie, z.B:

    '-'          solid line style (default)
    '--'         dashed line style
    '-.'         dash-dot line style
    ':'          dotted line style

````python
X = [1, 8, 7, 3]
Y = [3, 10, 2, 9]
plt.figure(figsize=(4,3))      
plt.plot(X, Y, color='red', linestyle='--', marker="o")
plt.show()
````

#### Beispiel:

Wir betrachten die Funktion $f(x) = x^3-3x^2-x+3$ mit ihrer Ableitung $f^{\prime}(x) = 3x^2 -6x -1$.

````python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 3*x**2 - x + 3

def f1(x):
    return 3*x**2 - 6*x - 1

X = np.linspace(-2.5, 5, 400)
Y = f(X)
Y1 = f1(X)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(X, Y, label='x**3 - 3*x**2 - x + 3', color="blue")
plt.plot(X, Y1, label='3*x**2 - 6*x - 1', color="red")

plt.xlim(-2,4)
plt.ylim(-5,4)

# Extremwerte
n1, n2 = -0.155, 2.155
plt.axvline(n1, color='green', linewidth=0.8, linestyle=':')
plt.axvline(n2, color='green', linewidth=0.8, linestyle=':')
plt.scatter([n1,n2],[f(n1),f(n2)])
plt.text(x=n1+0.2, y=f(n1), s=f'({n1:.2f}/{f(n1):.2f})')
plt.text(x=n2+0.2, y=f(n2), s=f'({n2:.2f}/{f(n2):.2f})')


# Beschriftung
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.title("Polynom mit Ableitung")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
````

#### Ticks

Der *MultipleLocator* plaziert Ticks an den Vielfachen der mitgegebenen Zahl. Der *AutoMinorLocator* erzeugt soviele gleichmäßige Unterteilungen, wie die mitgegebene Zahl angibt.

````python
def f(x):
    return x*np.exp(x)
````

````python
from matplotlib.ticker import MultipleLocator, AutoMinorLocator
X = np.linspace(-6,3,100)
Y = f(X)
fig, ax = plt.subplots(figsize=(8,5))
ax.axvline(0,color='black',linewidth=1.5)
ax.axhline(0,color='black',linewidth=1.5)

# major ticks
ax.yaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.tick_params(axis='x', which='major', labelsize=5)
ax.tick_params(axis='y', which='major', labelsize=9)

# minor ticks
ax.yaxis.set_minor_locator(AutoMinorLocator(3)) 
ax.xaxis.set_minor_locator(AutoMinorLocator(2)) 

plt.ylim(-1,5)
plt.xlim(-5,2)

plt.grid(True)
plt.plot(X,Y,label=r'$x\cdot e^{x}$')  
plt.legend()
plt.show()
````

#### Kurvenscharen


Eine Deich können wir mit der Funktion $f(x) = a x^2 \cdot e^{bx}$ modellieren.


````python
def deich(ax, a, b, von, bis):

    def f(x,a,b):
        return a*x**2*np.exp(b*x)

    a0 =  '' if a == 1 else str(a)
    b0 =  '' if b == 1 else str(b)
    label=a0+r'$x^2 \cdot e^{'+b0+'x}$'
    
    ax.axvline(0,color='black',linewidth=1)
    ax.axhline(0,color='black',linewidth=1)
    
    X = np.linspace(von,bis,500)
    Y = f(X,a,b)
    ax.plot(X,Y,label=label)  
    ax.grid(True)
    ax.legend()
````

````python
fig, axs = plt.subplots(1,2,figsize=(10,3))

deich(axs[0],5,0.5,-20,0)
deich(axs[0],5,0.4,-20,0)
deich(axs[0],5,0.3,-20,0)
deich(axs[1],5,0.5,-20,0)
deich(axs[1],4,0.5,-20,0)
deich(axs[1],3,0.5,-20,0)

plt.show()
````
