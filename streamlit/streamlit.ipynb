## Streamlit

#### Beispiel1

Streamlit verbietet die Änderung des session state keys eines Widgets nachdem das schon gerendert wurde.

Wenn man mit einem Button einen session state ändert und dann manuell ein st.rerun() aufruft, sollte man überlegen, ob nicht ein callback sinnvoller ist.

```python
import streamlit as st

st.title("Beispiel 1: Falsch (Fehler)")

st.write("""
Dieses Skript demonstriert den Fehler:
`StreamlitAPIException: ... cannot be modified after the widget ... is instantiated.`
""")

# Initialisierung
if "my_check" not in st.session_state:
    st.session_state.my_check = False

# 1. Das Widget wird HIER instanziiert.
# Streamlit merkt sich: "my_check" wurde in diesem Run schon verwendet.
st.checkbox("Ich bin eine Checkbox", key="my_check")

# 2. Wir versuchen, den Wert zu ändern, NACHDEM das Widget schon gezeichnet wurde.
if st.button("Setze Checkbox auf True (Fehler)"):
    # FEHLER: Wir ändern den State eines Widgets, das in diesem Run schon existiert.
    st.session_state.my_check = True
    st.rerun()
```

```python
import streamlit as st

st.title("Beispiel 2: Richtig (Callback)")

st.write("""
Hier wird der Fehler vermieden, indem ein **Callback** verwendet wird.
Der Callback wird ausgeführt, BEVOR das Skript erneut durchläuft.
""")

# Initialisierung
if "my_check_correct" not in st.session_state:
    st.session_state.my_check_correct = False

def set_true_callback():
    # Diese Funktion läuft VOR dem nächsten Rerun.
    # Zu diesem Zeitpunkt ist noch kein Widget gezeichnet worden (im neuen Run).
    st.session_state.my_check_correct = True

# 1. Das Widget wird instanziiert.
st.checkbox("Ich bin eine Checkbox", key="my_check_correct")

# 2. Der Button ruft den Callback auf.
# Wenn man klickt, führt Streamlit erst 'set_true_callback' aus,
# und startet DANN das Skript von oben neu.
st.button("Setze Checkbox auf True (Richtig)", on_click=set_true_callback)
```
